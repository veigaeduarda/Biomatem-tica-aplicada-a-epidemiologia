```python
#Utilizaremos o método SIR (Susceptíveis-Infectados-Recuperados)

# Valores iniciais
S0 = 203062512
I0 = 2298
R0 = 1431

# Parâmetros
beta = 0.0001  # Taxa de transmissão
gamma = 0.1    # Taxa de recuperação

# Período de simulação de 10 anos (em dias)
t_max = 3650  # 10 anos = 3650 dias
h = 1         # Passo de tempo (um dia)

# Listas para armazenar os resultados
S, I, R = [S0], [I0], [R0]
```

```python
def sir_model(S0, I0, R0, beta, gamma, t_max, h):
    S, I, R = [S0], [I0], [R0]
    t = 0

    while t < t_max:
        dS = -beta * S[-1] * I[-1]
        dI = beta * S[-1] * I[-1] - gamma * I[-1]
        dR = gamma * I[-1]

        k1_S = h * dS
        k1_I = h * dI
        k1_R = h * dR

        k2_S = h * (-beta * (S[-1] + k1_S/2) * (I[-1] + k1_I/2))
        k2_I = h * (beta * (S[-1] + k1_S/2) * (I[-1] + k1_I/2) - gamma * (I[-1] + k1_I/2))
        k2_R = h * gamma * (I[-1] + k1_I/2)

        k3_S = h * (-beta * (S[-1] + k2_S/2) * (I[-1] + k2_I/2))
        k3_I = h * (beta * (S[-1] + k2_S/2) * (I[-1] + k2_I/2) - gamma * (I[-1] + k2_I/2))
        k3_R = h * gamma * (I[-1] + k2_I/2)

        k4_S = h * (-beta * (S[-1] + k3_S) * (I[-1] + k3_I))
        k4_I = h * (beta * (S[-1] + k3_S) * (I[-1] + k3_I) - gamma * (I[-1] + k3_I))
        k4_R = h * gamma * (I[-1] + k3_I)

        S.append(S[-1] + (k1_S + 2*k2_S + 2*k3_S + k4_S)/6)
        I.append(I[-1] + (k1_I + 2*k2_I + 2*k3_I + k4_I)/6)
        R.append(R[-1] + (k1_R + 2*k2_R + 2*k3_R + k4_R)/6)

        t += h

    return S, I, R
```

```python
# Plot dos resultados
plt.plot(range(t_max + 1), S, label='Suscetíveis')
plt.plot(range(t_max + 1), I, label='Infectados')
plt.plot(range(t_max + 1), R, label='Recuperados')
plt.xlabel('Dias')
plt.ylabel('Número de Indivíduos')
plt.legend()
plt.title('Modelo SIR ao longo de 10 anos')
plt.show()
```
